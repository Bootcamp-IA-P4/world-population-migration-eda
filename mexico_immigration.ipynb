## EDA Migración de México debido al narcotráfico

```python
import pandas as pd
import numpy as np

# Importamos librerias para visualización de datos
import matplotlib.pyplot as plt
import seaborn as sns

# Configuramos las gráficas
plt.style.use('seaborn-v0_8') 

# Configurar pandas para mostrar números completos sin notación científica
pd.set_option('display.float_format', '{:.0f}'.format)
```

```python
df_world = pd.read_csv('data/world_population.csv')
print(f"El DataFrame World Population tiene {df_world.shape[0]} filas y {df_world.shape[1]} columnas.")

df_crime = pd.read_csv('data/crime.csv')
print(f"El DataFrame Mexico crime rate statistics  tiene {df_crime.shape[0]} filas y {df_crime.shape[1]} columnas.")

df_immigration = pd.read_csv('data/mexico_immigration.csv')
print(f"El DataFrame mexico immigration tiene {df_immigration.shape[0]} filas y {df_immigration.shape[1]} columnas.")
```
