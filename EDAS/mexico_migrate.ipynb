## 2. Procesamiento y Transformación de Datos

### 2.1. Conversión a formato Largo y ordenar por países Dataset Migración y Crímen

#### 2.1.1. Primero eliminamos columnas que no necesitamos

        #### Migración México

```python
# Migración
columnas_a_eliminar = ['Country Code', 'Indicator Name', 'Indicator Code', 'Unnamed: 68']

# Crear una copia del DataFrame sin esas columnas
df_immigration_clean = df_immigration.drop(columns=columnas_a_eliminar)

# Verificar las columnas restantes
print("\nColumnas después de eliminar:")
print(df_immigration_clean.columns.tolist())
```

        #### Crimen en México (Narcotráfico)

```python
# Crimen
columnas_a_eliminar = ['Country Code', 'Indicator Name', 'Indicator Code', 'Unnamed: 68']

# Crear una copia del DataFrame sin esas columnas
df_crime_clean = df_crime.drop(columns=columnas_a_eliminar)

# Verificar las columnas restantes
print("\nColumnas después de eliminar:")
print(df_crime_clean.columns.tolist())
```

### 2.1.2 Procedemos con la conversión a formato largo y ordenamos

#### Mexico Migration

```python
df_longm = pd.melt(
    df_immigration_clean,
    id_vars=['Country Name'],
    var_name='Year',
    value_name='Net_Migration'
)

# Ordenar el dataframe por país y año
df_longm = df_longm.sort_values(['Country Name', 'Year'])

# Resetear los índices
df_longm = df_longm.reset_index(drop=True)

# Mostrar las primeras filas para verificar la transformación
print("Primeras filas del DataFrame de Migración:")
print(df_longm.head(10))
```

#### Crime Rating 

```python
df_longcrime = pd.melt(
  df_crime_clean,
  id_vars=['Country Name'], 
  var_name='Year',
  value_name='Intentional homicides (per 100,000 people)'
)

# Ordenar el dataframe primero por país y luego por año
df_longcrime = df_longcrime.sort_values(by=['Country Name', 'Year'])

# Resetear los índices para que sean consecutivos
df_longcrime = df_longcrime.reset_index(drop=True)

# Mostrar las primeras filas para verificar la transformación inicial
print("Primeras filas del DataFrame de migración transformado:")
print(df_longcrime.head(10))
```

### 2.2 Conversion de tipo de datos datasets

### Migration

```python
# Verificar el tipo de datos actual de la columna 'Year'
print(f"Tipo de datos de la columna 'Year': {df_longm['Year'].dtype}")

#Convertir el año a tipo numérico
df_longm['Year'] = pd.to_numeric(df_longm['Year'])

# Verificar que la conversión fue exitosa
print(f"Nuevo tipo de datos de la columna 'Year': {df_longm['Year'].dtype}")

# Mostrar algunos datos
df_longm.head(5)
```

#### Crime

```python
# Verificar el tipo de datos actual de la columna 'Year'
print(f"Tipo de datos de la columna 'Year': {df_longcrime['Year'].dtype}")

#Convertir el año a tipo numérico
df_longcrime['Year'] = pd.to_numeric(df_longcrime['Year'])

# Verificar que la conversión fue exitosa
print(f"Nuevo tipo de datos de la columna 'Year': {df_longcrime['Year'].dtype}")

# Mostrar algunos datos
df_longcrime.head(5)
```

## 3. Crear un nuevo DataFrame combinando ambos

#### Migration - World Population

```python
#Se combinan los dos dataframes utilizando innerjoin
df_combinedmp = pd.merge(
    df_world,
    df_longm [['Country Name', 'Year', 'Net_Migration']],
)

# Verificar el resultado
print("DataFrame Migration combinado:")
print(df_combinedmp.head(10))
```

#### Population - Crime 

```python
#Combinación de crímen y world population
df_combinedpc = pd.merge(
    df_world,
    df_longcrime [['Country Name', 'Year', 'Intentional homicides (per 100,000 people)']],
)

# Verificar el resultado
print("DataFrame Migration y Crime combinado:")
print(df_combinedpc.head(10))
```

#### Crime - migration

```python
# Combinamos los DataFrames en función de 'Country Name' y 'Year'
df_combinedcm = pd.merge(
    df_longcrime [['Country Name', 'Year', 'Intentional homicides (per 100,000 people)']],
    df_longm[['Country Name', 'Year', 'Net_Migration' ]],
)

# Mostramos las primeras filas del DataFrame combinado
print("Primeras filas del DataFrame combinado:")
print(df_combinedcm.head(10))
```

#### Crime, Population & Migration

```python
# Combinar Crime, Population y Migration
df_all_combined = pd.merge(
    df_combinedcm,
    df_world [['Country Name', 'Year', 'Population']],
)

print("Primeras filas del DataFrame combinado:")
print(df_all_combined.head(10))
```

## 5. Análisis Migratorio en México basado en Violencia Criminal (Homicidios 2000 - 2023)

```python
mexico = df_combinedcm[df_combinedcm['Country Name'] == 'Mexico']

#Filtrar el período de 2000 a 2023
guerra_narco = mexico[(mexico['Year'] >= 1999) & (mexico['Year'] <= 2021)]

#Verificar el periodo correcto
print("Datos de México de 2000 a 2023:")
print(guerra_narco.head(20))
```

## 6. Analisis migratiorio basado en violencia criminal

#### 6.1 Tabla de visualizaci{on}

```python
# Verificar los valores negativos en Net_Migration
negativos = guerra_narco[guerra_narco['Net_Migration'] < 0]
print("Valores negativos en Net_Migration:")
print(negativos)

# Visualizar la tendencia de Net_Migration a lo largo de los años
plt.figure(figsize=(10, 6))
sns.lineplot(data=guerra_narco, x='Year', y='Net_Migration', marker='o')
plt.title('Tendencia de la Migración Neta en México (1999-2021)')
plt.xlabel('Año')
plt.ylabel('Migración Neta')
plt.grid(True)
plt.show()
```
