## 2. Procesamiento y Transformación de Datos

### 2.1. Conversión a formato Largo y ordenar por países Dataset Migración y Crímen

#### 2.1.1. Primero eliminamos columnas que no necesitamos

        #### Migración México

```python
# Migración
columnas_a_eliminar = ['Country Code', 'Indicator Name', 'Indicator Code', 'Unnamed: 68']

# Crear una copia del DataFrame sin esas columnas
df_immigration_clean = df_immigration.drop(columns=columnas_a_eliminar)

# Verificar las columnas restantes
print("\nColumnas después de eliminar:")
print(df_immigration_clean.columns.tolist())
```

        #### Crimen en México (Narcotráfico)

```python
# Crimen
columnas_a_eliminar = ['Country Code', 'Indicator Name', 'Indicator Code', 'Unnamed: 68']

# Crear una copia del DataFrame sin esas columnas
df_crime_clean = df_crime.drop(columns=columnas_a_eliminar)

# Verificar las columnas restantes
print("\nColumnas después de eliminar:")
print(df_crime_clean.columns.tolist())
```

### 2.1.2 Procedemos con la conversión a formato largo y ordenamos

#### Mexico Migration

```python
df_longm = pd.melt(
    df_immigration_clean,
    id_vars=['Country Name'],
    var_name='Year',
    value_name='Net_Migration'
)

# Ordenar el dataframe por país y año
df_longm = df_longm.sort_values(['Country Name', 'Year'])

# Resetear los índices
df_longm = df_longm.reset_index(drop=True)

# Mostrar las primeras filas para verificar la transformación
print("Primeras filas del DataFrame de Migración:")
print(df_longm.head(10))
```

#### Crime Rating 

```python
df_longcrime = pd.melt(
  df_crime_clean,
  id_vars=['Country Name'], 
  var_name='Year',
  value_name='Intentional homicides (per 100,000 people)'
)

# Ordenar el dataframe primero por país y luego por año
df_longcrime = df_longcrime.sort_values(by=['Country Name', 'Year'])

# Resetear los índices para que sean consecutivos
df_longcrime = df_longcrime.reset_index(drop=True)

# Mostrar las primeras filas para verificar la transformación inicial
print("Primeras filas del DataFrame de migración transformado:")
print(df_longcrime.head(10))
```

### 2.2 Conversion de tipo de datos datasets

### Migration

```python
# Verificar el tipo de datos actual de la columna 'Year'
print(f"Tipo de datos de la columna 'Year': {df_longm['Year'].dtype}")

#Convertir el año a tipo numérico
df_longm['Year'] = pd.to_numeric(df_longm['Year'])

# Verificar que la conversión fue exitosa
print(f"Nuevo tipo de datos de la columna 'Year': {df_longm['Year'].dtype}")

# Mostrar algunos datos
df_longm.head(5)
```

#### Crime

```python
# Verificar el tipo de datos actual de la columna 'Year'
print(f"Tipo de datos de la columna 'Year': {df_longcrime['Year'].dtype}")

#Convertir el año a tipo numérico
df_longcrime['Year'] = pd.to_numeric(df_longcrime['Year'])

# Verificar que la conversión fue exitosa
print(f"Nuevo tipo de datos de la columna 'Year': {df_longcrime['Year'].dtype}")

# Mostrar algunos datos
df_longcrime.head(5)
```

## 3. Crear un nuevo DataFrame combinando ambos

#### Migration - World Population

```python
#Se combinan los dos dataframes utilizando innerjoin
df_combinedmp = pd.merge(
    df_world,
    df_longm [['Country Name', 'Year', 'Net_Migration']],
)

# Verificar el resultado
print("DataFrame Migration combinado:")
print(df_combinedmp.head(10))
```

#### Population - Crime 

```python
#Combinación de crímen y world population
df_combinedpc = pd.merge(
    df_world,
    df_longcrime [['Country Name', 'Year', 'Intentional homicides (per 100,000 people)']],
)

# Verificar el resultado
print("DataFrame Migration y Crime combinado:")
print(df_combinedpc.head(10))
```

#### Crime - migration

```python
# Combinamos los DataFrames en función de 'Country Name' y 'Year'
df_combinedcm = pd.merge(
    df_longcrime [['Country Name', 'Year', 'Intentional homicides (per 100,000 people)']],
    df_longm[['Country Name', 'Year', 'Net_Migration' ]],
)

# Mostramos las primeras filas del DataFrame combinado
print("Primeras filas del DataFrame combinado:")
print(df_combinedcm.head(10))
```

#### Crime, Population & Migration

```python
# Combinar Crime, Population y Migration
df_all_combined = pd.merge(
    df_combinedcm,
    df_world [['Country Name', 'Year', 'Population']],
)

print("Primeras filas del DataFrame combinado:")
print(df_all_combined.head(10))
```

## 5. Análisis Migratorio en México basado en Violencia Criminal (Homicidios 2000 - 2023)

```python
mexico = df_combinedcm[df_combinedcm['Country Name'] == 'Mexico']

#Filtrar el período de 2000 a 2023
guerra_narco = mexico[(mexico['Year'] >= 1980) & (mexico['Year'] <= 2025)]

#Verificar el periodo correcto
print("Datos de México de 2000 a 2023:")
print(guerra_narco.head(20))
```

## 6. Analisis migratiorio basado en violencia criminal

#### 6.1 Tabla de visualizaci{on}

```python
# Verificar los valores negativos en Net_Migration
negativos = guerra_narco[guerra_narco['Net_Migration'] < 0]
print("Valores negativos en Net_Migration:")
print(negativos)

# Visualizar la tendencia de Net_Migration a lo largo de los años
plt.figure(figsize=(10, 6))
sns.lineplot(data=guerra_narco, x='Year', y='Net_Migration', marker='o')
plt.title('Tendencia de la Migración Neta en México (1999-2021)')
plt.xlabel('Año')
plt.ylabel('Migración Neta')
plt.grid(True)
plt.show()
```

#### 6.2 Correlación de crimen con migración

```python
# Calcular correlación
correlacion = guerra_narco[['Net_Migration', 'Intentional homicides (per 100,000 people)']].corr()
print(f"Correlación entre Migración Neta y Homicidios Intencionales: ")
print(correlacion)

# Extraer el coeficiente específico de correlación
coef_corr = correlacion.iloc[0, 1]  # Valor en la posición [0,1] de la matriz
print(f"Coeficiente de correlación: {coef_corr:.3f}")

# Matriz de correlación con todas las variables numéricas
matriz_corr = guerra_narco.select_dtypes(include='number').corr()
print("\nMatriz de correlación completa:")
print(matriz_corr)
```

#### 6.3. Visualización de la gráfica de la correlación

```python
# Gráfico de dispersión para mostrar la relación
plt.figure(figsize=(12, 8))

# Crear el gráfico de dispersión
scatter = sns.scatterplot(data=guerra_narco, 
                x='Intentional homicides (per 100,000 people)', 
                y='Net_Migration')

# Añadir línea de tendencia
sns.regplot(data=guerra_narco, 
            x='Intentional homicides (per 100,000 people)', 
            y='Net_Migration',
            scatter=False, 
            line_kws={"color":"red"})

# Crear un diccionario para eventos importantes
# Será necesario localizar las coordenadas (tasa de homicidios, migración neta) para cada año
eventos = {
    # Terremotos
    2017: "Terremoto 2017",
    
    # Cambios de gobierno
    2006: "Inicio Felipe Calderón\nGuerra contra narco", 
    2012: "Inicio Peña Nieto", 
    2018: "Inicio AMLO"
}

# Añadir anotaciones para cada evento
for year, label in eventos.items():
    if year in guerra_narco['Year'].values:
        # Obtener los valores x,y para este año
        punto = guerra_narco[guerra_narco['Year'] == year]
        x = punto['Intentional homicides (per 100,000 people)'].values[0]
        y = punto['Net_Migration'].values[0]
        
        # Añadir anotación
        plt.annotate(label, 
                     xy=(x, y),
                     xytext=(10, 10),
                     textcoords='offset points',
                     arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.2'),
                     bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.7))

# Añadir eventos fuera del rango temporal como nota
plt.figtext(0.5, 0.01, 
           "Nota: El terremoto de 1985 y la crisis económica de los 90s están fuera del rango temporal de los datos (1999-2021)",
           ha="center", fontsize=10, bbox={"facecolor":"orange", "alpha":0.2, "pad":5})

plt.title('Relación entre Tasa de Homicidios y Migración Neta')
plt.xlabel('Homicidios Intencionales (por 100,000 habitantes)')
plt.ylabel('Migración Neta')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Crear dos ejes Y para visualizar ambas variables
fig, ax1 = plt.subplots(figsize=(15, 8))

# Eje Y1: Migración Neta
color = 'tab:blue'
ax1.set_xlabel('Año')
ax1.set_ylabel('Migración Neta', color=color)
ax1.plot(guerra_narco['Year'], guerra_narco['Net_Migration'], color=color, marker='o')
ax1.tick_params(axis='y', labelcolor=color)

# Eje Y2: Tasa de Homicidios
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Homicidios (por 100,000)', color=color)
ax2.plot(guerra_narco['Year'], guerra_narco['Intentional homicides (per 100,000 people)'], 
         color=color, marker='s')
ax2.tick_params(axis='y', labelcolor=color)

# Eventos importantes: cambios presidenciales y terremotos
eventos = {
    # Presidentes desde 1980
    1985: "Terremoto CDMX",
    1988: "Salinas",
    1994: "Zedillo\nCrisis del 94",
    2000: "Fox",
    2006: "Calderón\nGuerra contra narco",
    2012: "Peña Nieto",
    2017: "Terremoto\nCDMX/Chiapas",
    2018: "AMLO",
    2020: "Pandemia\nCOVID-19"
}

# Añadir líneas verticales y etiquetas para cada evento
min_y1, max_y1 = ax1.get_ylim()
altura_texto = min_y1 + (max_y1 - min_y1) * 0.9  # Posición para las etiquetas

for year, label in eventos.items():
    if year in guerra_narco['Year'].values or (year >= min(guerra_narco['Year']) and year <= max(guerra_narco['Year'])):
        # Verificar si el año está cerca de algún año en los datos (para evitar errores)
        # Añadir línea vertical punteada
        ax1.axvline(x=year, color='green', linestyle='--', alpha=0.7)
        
        # Añadir etiqueta
        ax1.text(year, altura_texto, label, rotation=90, 
                verticalalignment='top', horizontalalignment='right',
                bbox=dict(facecolor='yellow', alpha=0.5, pad=3))

# Añadir leyenda para las líneas
ax1.plot([], [], color='tab:blue', marker='o', label='Migración Neta')
ax1.plot([], [], color='tab:red', marker='s', label='Tasa de Homicidios')
ax1.legend(loc='upper left')

# Ajustar título para reflejar el rango completo de años
min_year = min(guerra_narco['Year'])
max_year = max(guerra_narco['Year'])
plt.title(f'Tendencias de Migración Neta y Homicidios en México ({min_year}-{max_year})\ncon eventos históricos importantes', fontsize=14)

# Mejorar visual
plt.grid(True, alpha=0.3)
fig.tight_layout(rect=[0, 0.03, 1, 0.97])  # Dejar espacio para la nota

plt.show()
```
